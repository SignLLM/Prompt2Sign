```python
import random

# Read file contents, the input.text is the line text of the video
with open('./example_input.text', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# List of prompt words (default English, You can use German, prompts_GSL.txt)
with open('prompts_ASL.txt', 'r', encoding='utf-8') as file:
    prompts = [line.strip() for line in file.readlines()]

def transform_sentence(sentence):
    """
    Insert the sentence into a randomly selected template.
    """
    prompt_text = random.choice(prompts).format(sentence.strip())
    return prompt_text

# Sentence by sentence conversion
new_sentences = [transform_sentence(line) for line in lines]

# Write to new file
with open('./example_output.text', 'w', encoding='utf-8') as f:
    for sentence in new_sentences:
        f.write(sentence + '\n')
```
